```python
import spacy
from negspacy.negation import Negex
import pandas as pd
```

```python
nlp = spacy.load("en_core_web_sm")
nlp.add_pipe("negex")
```

```python
def negex_text(s):
    doc = nlp(s)
    return [(e.text, e._.negex) for e in doc.ents]
```

```python
# Read in data
df = pd.read_csv("/content/drive/MyDrive/fake-news-explainability/Data/fake_news_train.csv")
df = df.dropna(subset=['text']).reset_index(drop=True)
df = df[['id','text','label']]

# Find statements with has, is, will, not
df = df[df['text'].str.contains(" has | is | will | was ")].reset_index(drop=True)
```

```python
df['negex'] = df['text'].apply(negex_text)
```
